```
# 답안예시 코드
input_data = input()
row = int(input_data[1])
column = int(ord(input_data[0])) - int(ord('a')) + 1

steps = [(-2, -1),(-1,-2),(1,-2),(2,-1),(2,1),(1,2),(-1,2),(-2,1)]

cnt = 0
for step in steps:
  next_row = row + step[0]
  next_col = column + step[1]

  if next_row >= 1 and next_row <= 8 and  next_col >= 1 and  next_col <= 8:
    cnt +=1

print(cnt)
```

```
# 답안예시 코드
n, m = map(int,input().split())

d = [[0] * m for _ in range(n)]

x, y , start = map(int,input().split())
d[x][y] = 1

array = []
for i in range(n):
  array.append(list(map(int, input().split())))

dx = [-1, 0, 1, 0]
dy = [0, 1, 0, -1]

def turn_left():
  global start
  start -= 1
  if start == -1:
    start = 3

cnt = 1
turn_time = 0
while True:
  turn_left()
  nx = x + dx[start]
  ny = y + dy[start]

  if d[nx][ny] == 0 and array[nx][ny] == 0:
    d[nx][ny] = 1
    x = nx
    y = ny
    cnt +=1
    turn_time = 0
    continue
  else:
    turn_time += 1
  if turn_time == 4:
    nx = x - dx[start]
    ny = y - dy[start]
    if array[nx][ny] == 0 :
      x = nx
      y = ny
    else:
      break
    turn_time = 0
print(cnt)
```
